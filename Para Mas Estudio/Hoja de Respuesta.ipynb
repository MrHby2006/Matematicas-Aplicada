7) El tamaño de una base de datos en terabytes (TB) crece exponencialmente en función del tiempo en años, a través del siguiente modelo:

﻿D(t) = 2 X 1,2 elevado a t﻿



¿Cuántos años deben pasar que la base de datos tenga un tamaño de 20 TB?

```python
import numpy as np
from scipy.optimize import fsolve
def D(t):
    return 2 * 1.2**t - 20

xo = np.linspace(0, 20, 1)
solucion = fsolve(D, xo)

print(f"{solucion}")
```

13) En una tienda recién inaugurada, la primera semana ha vendido 300 unidades de un producto determinado y se espera que las ventas aumenten en 40 unidades cada semana. ¿Cuántas unidades del producto se proyecta vender en total al cabo de 2 meses? Considere que 1 mes equivale a 4 semanas.

```python
a_1 = 300
d = 40
n = 8
sucesion = []
for i in range(8):
    n = i + 1
    sucesion.append(a_1 + (n - 1) * d)

suma = sum(sucesion)
print(f"{suma}")
```

19) La matriz ﻿A igual abrir paréntesis a subíndice i j fin subíndice cerrar paréntesis﻿ nos indica la cantidad de acciones tipo ﻿j﻿, pertenecientes a una empresa, que posee el inversionista ﻿i﻿.

Por otra parte, la matriz ﻿E igual abrir paréntesis e subíndice i j fin subíndice cerrar paréntesis﻿ nos detalla el valor unitario de la acción tipo ﻿i﻿ de esta empresa, cotizado en euros, durante el día hábil ﻿j﻿ de la semana pasada.



Las matrices ﻿A﻿ y ﻿E﻿ tienen orden ﻿15 x 10﻿ y ﻿10 x 5﻿ respectivamente y se definen por las siguientes expresiones:



﻿a subíndice i j fin subíndice igual 20 por i j﻿



﻿e subíndice i j fin subíndice igual 4 por i más j﻿



¿En cuánto está evaluado, para el miércoles de la semana pasada, el total de acciones que posee en esta empresa el inversionista 8?

```python
import numpy as np

A = np.fromfunction(lambda i, j: 20*(i + 1) * (j + 1), (15, 10))
E = np.fromfunction(lambda i, j: 4*(i + 1) + (j + 1), (10, 5))

C = np.dot(A, E)

print(f"{C}")
```

20) GB de RAM y 6 unidades de almacenamiento. Si la empresa ha adquirido un total de 72 GB de RAM y 40 unidades de almacenamiento, el sistema de ecuaciones que permite determinar la cantidad de estaciones de trabajo y servidores que se han instalado es:



﻿abrir tabla fila celda 4 x más 8 y igual 72 fin celda fila celda 2 x más 6 y igual 40 fin celda fin tabla cerrar llaves﻿



¿Cuántas estaciones de trabajo y servidores se han comprado?

```python
A = np.array([[4, 8],
              [2, 6]])

B = np.array([[72],
              [40]])

A_inv = np.linalg.inv(A)

C = np.dot(A_inv, B)

print(f"{C}")
```
